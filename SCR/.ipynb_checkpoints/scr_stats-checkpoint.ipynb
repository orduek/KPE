## Frequentist analysis

```python
## Start with simple GLM models (using statsmodel)
glm = smf.mixedlm('SCR ~ time * groupIdx', data=df_long, groups = df_long.scr_id, 
                 missing='drop')
model1 = glm.fit()
print(model1.summary())
```

```python
pg.mixed_anova(data=df_long, dv='SCR', within='time', between='groupIdx', subject='scr_id').round(3)
```

- Interaction between group and time, suggesting differences between the groups on the second time
- Move on to Bayes

## Go Bayes

```python
## set indexes
# set all the rest
df_long['scr_id'] = df_long.scr_id.astype('category')
sub_idx = df_long.scr_id.cat.codes.values # get county category index values (613 x 1) vec with nums [0-7]
print(sub_idx)
n_subs = len(np.unique(df_long.scr_id))
print(n_subs)
# set time as index also
#df_clean.groupIdx = pd.Categorical(df_clean.groupIdx)
df_long.time = pd.Categorical(df_long.time, 
                    categories=['beforeTrt','afterTrt','FU30'])
timeIDX = df_long.time.cat.codes.values
print(timeIDX)
```

```python
with pm.Model() as model1:
 
    a = pm.Normal('a', mu=0, sd=1, shape=n_subs)
       
    b = pm.Normal('b', mu=0, sd=1, shape=3) # time
    
    bMed = pm.Normal('bMed', mu=0, sd=.5) # two groups
  
    # Model error
    eps = pm.HalfCauchy('eps', 3)

    y_hat = a[sub_idx] + b[timeIDX] + bMed*df_long.groupIdx 
    # Data likelihood
    likelihood = pm.Normal('likelihood', mu=y_hat,
                           sigma=eps, observed=df_long.SCR)
    trace = pm.sample(draws=4000, tune=2000, target_accept=0.9, return_inferencedata=False)
```

```python
az.plot_trace(trace)
```

```python
az.summary(trace)
```

```python
# lets look at each time point
sns.distplot(trace["b"][:,0])
sns.distplot(trace["b"][:,1])
sns.distplot(trace["b"][:,2])
#sns.distplot(trace["b"][:,3])
```

```python
# lets look at group at each time point
with pm.Model() as model_glm:
    pm.GLM.from_formula('afterTrt ~ groupIdx', df)
    trace_after = pm.sample(draws=4000, tune=2000)
az.summary(trace_after, hdi_prob=.9)
```

### Plotting

```python
# set variables
def plotDiff(depVar,indepVar, trace):
    sns.set_style("ticks") # set style
    y = str(depVar)
    dfPlot = df
    ci = np.quantile(trace[str(indepVar)], [.05,.95])
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(3, 5),gridspec_kw={'width_ratios': [1, .2],
                                                            'wspace':.1})
    g1 = sns.stripplot(y= y, x='med_cond', data=dfPlot, size = 8, ax=ax1)
    sns.boxplot(y= y, x='med_cond', data=dfPlot,  ax=ax1,
                boxprops=dict(alpha=.3))
    g2 = sns.distplot(trace[str(indepVar)], ax = ax2, vertical=True)
    ax2.vlines(x=0.001,ymin=ci[0], ymax=ci[1], color='black', 
               linewidth = 3, linestyle = "-")

    #g3.set_ylim(-.7, .7)
    #ax1.set_ylim(-.7,.7)
    ax1.set_ylim(g2.get_ylim()) # use first graph's limits to get the relevant for this one
    ax2.yaxis.tick_right()
    ax2.set_xticks([])
    ax2.set_ylabel("Difference between groups", fontsize=14) 
    ax2.yaxis.set_label_position("right")
    ax1.set_ylabel("SCR to traumatic script", fontsize=12)
    ax1.set_xlabel("Group", fontsize=14)
    fig.savefig( str(depVar) + 'SCR.png', dpi=300, bbox_inches='tight')
    return fig
```

```python
ax1 = plotDiff('afterTrt','groupIdx',trace_after)
```
